```python
def mean_absolute_percentage_error(y_val, y_pred): 
    y_val, y_pred = np.array(y_val), np.array(y_pred)
    return np.mean(np.abs((y_val - y_pred) / y_val)) * 100
```

```python
from sklearn.metrics import r2_score

def r_square(y_actual, y_prediction):
    return(r2_score(y_actual, y_prediction)*100)
```
